### Hypobolic functions
The reason why it is called hypobolic sine /hypobolic cosine is due to the analogs of the circular function, since $(\cos t, \sin t)$ forms a circle $x^2 + y^2 = 1$. So it is defined artificially that the points $(\cosh t, \sinh t)$ forms a hyperbola $x^2 - y^2 = 1$

$$\sinh x = \frac{e^x - e^{-x}}{2}$$
$$\cosh x = \frac{e^x + e^{-x}}{2}$$
$$\tanh x = \frac{\sinh x}{\cosh x}$$
$$\coth x = \frac{\cosh x}{\sinh x}$$

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(-5, 5, 100)
tanh = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
coth = (np.exp(x) + np.exp(-x))/(np.exp(x) - np.exp(-x))
sinh = (np.exp(x) - np.exp(-x))/2
cosh = (np.exp(x) + np.exp(-x))/2


fig, ax = plt.subplots()

ax.plot(x, tanh, label='tanh')
ax.plot(x, sinh, label='sinh')
ax.plot(x, cosh, label='cosh')
# ax.plot(x, coth, label='coth')
ax.plot(x, np.exp(x), label='$exp(x)$')
ax.legend()
plt.axhline(y=0, linestyle='--', c='k')
plt.axvline(x=0, linestyle='--', c='k')
```

### Fly Straight Dammit

plots of the graph introduced by numberphile: [Amazing Graphs](https://www.youtube.com/watch?v=pAMgUB51XZA)

```python
def gcd(x, y):
    if x < y: 
        small = x
    else: 
        small = y

    gcd = 0
    for i in range(1, int(small)+1):
        if x%i == 0 and y%i == 0: 
            if gcd < i:
                gcd = i

    return gcd   
```

```python
x = np.arange(0, 1000)
y = [1, 1]
for i in range(2, len(x)):
    if gcd(x[i], y[-1]) == 1: 
        assert y[-1] == y[i-1], "wrong index matching"
        y.append(x[i] + y[-1] + 1)
    elif gcd(x[i], y[-1]) > 1: 
        y.append(y[-1]//gcd(x[i], y[-1]))
    else:
        raise ValueError()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

### Parallelogram

plots of a graph with the points forming the parallelogram, introduced by [Amazing Graphs - Numberphile](https://www.youtube.com/watch?v=pAMgUB51XZA)

```python
def isprime(x):
    for i in range(1, x+1):
        if x%i == 0 and i > 1 and i != x:
            return False
    return True
```

```python
max_x = 10000
x = []
for i in range(2, max_x):
    if isprime(i):
        x.append(i)
y = []
for i in x:
    ori_bin = bin(i)[2:]
    reverse_bin = list(str(ori_bin))[::-1]
    s_reverse_bin = ""
    for j in reverse_bin: 
        s_reverse_bin += j
    reverse_dec = int(s_reverse_bin, 2)
    result = i - reverse_dec
    y.append(result)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```
